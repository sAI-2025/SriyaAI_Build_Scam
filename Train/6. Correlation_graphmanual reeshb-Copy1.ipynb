```python
def adjust_badclasscount(badclasscount, perchange):

    return badclasscount - (badclasscount * perchange)


badclasscount = 591  # Original bad class count
perchange = .8 # 10% change
adjusted_count = adjust_badclasscount(badclasscount, perchange)

print(f"Original badclasscount: {badclasscount}")
print(f"Adjusted badclasscount after {perchange*100}% change: {round(adjusted_count)}")
```
