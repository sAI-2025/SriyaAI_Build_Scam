```python
def show_mismatches(df1, df2, name1, name2):
    # Align both DataFrames by primary_key
    df1_aligned = df1.sort_values("primary_key").reset_index(drop=True)
    df2_aligned = df2.sort_values("primary_key").reset_index(drop=True)

    if not df1_aligned.equals(df2_aligned):
        print(f"\n❌ Mismatched rows between {name1} and {name2}:")
        try:
            mismatches = df1_aligned.compare(df2_aligned, align_axis=0)
            print(mismatches.head(10))  # Show first 10 mismatches
        except Exception as e:
            print("⚠️ Error during comparison:", str(e))
    else:
        print(f"\n✅ {name1} and {name2} are completely identical on selected columns.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_class_distribution(df, target_col, feature_col, kind='auto'):
    """
    Plots the distribution of a feature column split by classes in the target column.

    Parameters:
        df (pd.DataFrame): The input dataframe.
        target_col (str): The name of the target/class column.
        feature_col (str): The column whose distribution you want to plot.
        kind (str): 'auto', 'hist', 'kde', or 'count' — optional.
    """
    plt.figure(figsize=(10, 6))

    # Determine type of plot automatically if needed
    if kind == 'auto':
        if pd.api.types.is_numeric_dtype(df[feature_col]):
            kind = 'kde'  # or 'hist' if preferred
        else:
            kind = 'count'

    if kind in ['hist', 'kde']:
        for cls in df[target_col].unique():
            subset = df[df[target_col] == cls]
            sns.kdeplot(
                data=subset,
                x=feature_col,
                label=f'{target_col} = {cls}',
                fill=True,
                common_norm=False,
                alpha=0.5
            ) if kind == 'kde' else sns.histplot(
                data=subset,
                x=feature_col,
                label=f'{target_col} = {cls}',
                bins=30,
                kde=False,
                alpha=0.5
            )
        plt.title(f'Distribution of {feature_col} by {target_col}')
        plt.xlabel(feature_col)
        plt.ylabel('Density' if kind == 'kde' else 'Count')

    elif kind == 'count':
        sns.countplot(data=df, x=feature_col, hue=target_col)
        plt.title(f'Count of {feature_col} by {target_col}')
        plt.xlabel(feature_col)
        plt.ylabel('Count')

    else:
        raise ValueError("Invalid 'kind'. Use 'auto', 'hist', 'kde', or 'count'.")

    plt.legend(title=target_col)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
from yellowbrick.classifier import ConfusionMatrix,ClassificationReport,ROCAUC
def ConfusionMat(clf, X_train, y_train,X_test, y_test,classes,alpha):
    print('For alpha: ',alpha)
    cm = ConfusionMatrix(clf, classes=classes,title="Confusion Matrix")
    cm.fit(X_train, y_train)
    cm.score(X_test, y_test)
    cm.poof()
    cfm = cm.confusion_matrix_
    return cfm
```

```python
# ============================================================
# STEP 1: INITIALIZE ALL METRIC STORAGE LISTS
# ============================================================
predtst = []
gfforward = []
bfforward = []
totdxiac = []
gooddxicount = []
baddxicount = []
gddxacc = []
bddxacc = []
gdclwgt = []
prec, prec1 = [], []
rec, rec1 = [], []
bdclwgt = []
fps = []
fns = []
tps = []
accs, accs1 = [], []
tns = []
pgscr = []
pbscr = []
pgwt = []
pbwt = []
actsgd = []
actsbd = []
prdtsgd = []
prdtsbd = []
totprdacc = []
predTEST = []
actTEST = []
aucscores, aucscores1 = [], []
fprs, tprs = [], []

# Train, Validation, Test split storage
Xtest, Xtrain, ytrain, Xval, yval, predsval = [], [], [], [], [], []
ytest = []  # NEW: Store test labels

# Validation metrics
aucscoresval, fprsval, tprsval, accsval, precval, recval = [], [], [], [], [], []

# NEW: Additional metrics for comprehensive tracking
f1scores_val = []  # F1 scores for validation set
f1scores_test = []  # F1 scores for test set
predTESTproba = []  # Test probability predictions
predVALproba = []  # Validation probability predictions

# Model storage
sximod = []

# ============================================================
# STEP 2: ALPHA TUNING LOOP - ITERATE THROUGH DIFFERENT ALPHAS
# ============================================================
for i in range(len(alphas)):
    n = alphas[i]
    new_avg = n * avg_composite_dxi
    y_forward = create_label_forward(composite_dxi1, new_avg)
    print(f'\n{"="*60}')
    print(f'Processing Alpha: {n} (Iteration {i+1}/{len(alphas)})')
    print(f'{"="*60}')

    # ============================================================
    # STEP 3: CREATE LABELED DATASET
    # ============================================================
    df['gd_bdDXI'] = y_forward

    # Correctly Classified Good (True Negatives region)
    ccgd = df.loc[(df['gd_bdDXI'] == good) & (df[tv] == good)]

    # Correctly Classified Bad (True Positives region)
    ccbd = df.loc[(df['gd_bdDXI'] == bad) & (df[tv] == bad)]

    # Wrongly Classified Good (False Positives)
    wcgd = df.loc[(df['gd_bdDXI'] == good) & (df[tv] == bad)]

    # Wrongly Classified Bad (False Negatives)
    wcbd = df.loc[(df['gd_bdDXI'] == bad) & (df[tv] == good)]

    # Sample wrongly classified instances
    wcbda = wcbd.sample(n=round(len(wcbd) * 1), random_state=42)
    wcgda = wcgd.sample(n=round(len(wcgd) * 1), random_state=42)

    # Concatenate all samples
    cc = pd.concat([ccgd, ccbd, wcbda, wcgda], axis=0)
    cc.reset_index(inplace=True, drop=True)

    # ============================================================
    # STEP 4: PREPARE FEATURES AND TARGET
    # ============================================================
    X = cc.drop([tv, 'gd_bdDXI', 'composite_dxi_label', "primary_key"], axis=1)
    y = cc[tv]

    print(f'Total dataset size: {len(cc)}')
    print(f'Good class count: {len(ccgd)}')
    print(f'Bad class count: {len(ccbd)}')
    print(f'Class distribution: {y.value_counts().to_dict()}')

    # ============================================================
    # STEP 5: CHECK FOR EMPTY CLASSES (EDGE CASE HANDLING)
    # ============================================================
    if (len(ccbd) == 0) or (len(ccgd) == 0):
        print('⚠️ WARNING: One or both classes have zero samples. Skipping this alpha.')
        qwe = None
        # Append None to all metric lists
        accs.append(qwe)
        prec.append(qwe)
        rec.append(qwe)
        f1scores_test.append(qwe)  # NEW
        prdtsgd.append(qwe)
        prdtsbd.append(qwe)
        gfforward.append(qwe)
        bfforward.append(qwe)
        gooddxicount.append(qwe)
        baddxicount.append(qwe)
        gddxacc.append(qwe)
        bddxacc.append(qwe)
        gdclwgt.append(qwe)
        bdclwgt.append(qwe)
        totdxiac.append(qwe)
        pgscr.append(qwe)
        pbscr.append(qwe)
        pgwt.append(qwe)
        pbwt.append(qwe)
        totprdacc.append(qwe)
        actsgd.append(qwe)
        actsbd.append(qwe)
        fps.append(qwe)
        fns.append(qwe)
        tps.append(qwe)
        tns.append(qwe)
        Xtest.append(qwe)
        Xtrain.append(qwe)
        ytrain.append(qwe)
        ytest.append(qwe)  # NEW
        Xval.append(qwe)
        yval.append(qwe)
        predsval.append(qwe)
        predTEST.append(qwe)
        actTEST.append(qwe)
        aucscores.append(qwe)
        fprs.append(qwe)
        tprs.append(qwe)
        aucscoresval.append(qwe)
        fprsval.append(qwe)
        tprsval.append(qwe)
        accsval.append(qwe)
        precval.append(qwe)
        recval.append(qwe)
        f1scores_val.append(qwe)  # NEW
        predTESTproba.append(qwe)  # NEW
        predVALproba.append(qwe)  # NEW
        sximod.append(qwe)
        continue  # Skip to next alpha

    # ============================================================
    # STEP 6: TRAIN-VALIDATION-TEST SPLIT (70-10-20)
    # ============================================================
    from sklearn.model_selection import train_test_split

    # Assume: X and y are already defined
    # tv is the target variable (i.e., y = df['tv'])

    # First split: 80% (training + validation), 20% (test)
    non_test_X, X_test, non_test_y, y_test = train_test_split(
        X, y, stratify=y, test_size=0.2, random_state=42
    )

    # Second split: 10% of original (12.5% of 80%) goes to validation
    X_train, validate_X, y_train, validate_y = train_test_split(
        non_test_X, non_test_y, stratify=non_test_y, test_size=0.125, random_state=42
    )
      # Combine X and y for each set
    train_df = pd.concat([X_train, y_train], axis=1)
    val_df = pd.concat([validate_X, validate_y], axis=1)
    test_df = pd.concat([X_test, y_test], axis=1)

    train_df.to_csv(f"train_{i}.csv", index=False)
    val_df.to_csv(f"valid_{i}.csv", index=False)
    test_df.to_csv(f"test_{i}.csv", index=False)

    # Print split sizes
    print(f'Train size: {len(X_train)} ({len(X_train)/len(cc)*100:.1f}%)')
    print(f'Validation size: {len(validate_X)} ({len(validate_X)/len(cc)*100:.1f}%)')
    print(f'Test size (blind): {len(X_test)} ({len(X_test)/len(cc)*100:.1f}%)')

    # Function to print class distribution
    def print_class_distribution(y_split, name):
        print(f'\n{name} class distribution (%):')
        print(y_split.value_counts(normalize=True).mul(100).round(2))

    # Check if class distribution is preserved
    print_class_distribution(y, "Original")
    print_class_distribution(y_train, "Train")
    print_class_distribution(validate_y, "Validation")
    print_class_distribution(y_test, "Test")

    # Store splits
    Xtrain.append(X_train)
    ytrain.append(y_train)
    Xtest.append(X_test)
    ytest.append(y_test)
    Xval.append(validate_X)
    yval.append(validate_y)

    # ============================================================
    # STEP 7: TRAIN XGBOOST MODEL
    # ============================================================
    print(f'\n--- Training XGBoost Model ---')
    xgb = xg.XGBClassifier(random_state=42, eval_metric='logloss')
    xgb.fit(X_train.iloc[:, 1:], y_train)
    sximod.append(xgb)
    print('✓ Model training completed')

    # ============================================================
    # STEP 8: VALIDATION SET EVALUATION
    # ============================================================
    print(f'\n--- Validation Set Evaluation ---')

    # Predictions
    predval = xgb.predict(validate_X.iloc[:, 1:]).tolist()
    y_probsval = xgb.predict_proba(validate_X.iloc[:, 1:])[:, 1]

    # Store predictions
    predsval.append(predval)
    predVALproba.append(y_probsval.tolist())

    # Calculate metrics
    accval = accuracy_score(validate_y, predval)
    precival = precision_score(validate_y, predval, zero_division=0)
    recalval = recall_score(validate_y, predval, zero_division=0)
    f1val = f1_score(validate_y, predval, zero_division=0)

    # Store validation metrics
    accsval.append(accval)
    precval.append(precival)
    recval.append(recalval)
    f1scores_val.append(f1val)

    # Print validation metrics
    print(f'Validation Accuracy: {accval:.4f}')
    print(f'Validation Precision: {precival:.4f}')
    print(f'Validation Recall: {recalval:.4f}')
    print(f'Validation F1 Score: {f1val:.4f}')

    # Validation AUC-ROC
    fprval, tprval, thresholdsval = roc_curve(validate_y, y_probsval)
    roc_aucval = auc(fprval, tprval)
    aucscoresval.append(roc_aucval)
    fprsval.append(fprval)
    tprsval.append(tprval)
    print(f'Validation AUC Score: {roc_aucval:.4f}')

    # Plot Validation ROC Curve
    plt.style.use('ggplot')
    plt.figure(figsize=(10, 6), dpi=300)
    plt.plot(fprval, tprval, color='darkorange', lw=2, linestyle='--',
             label=f'Validation AUC = {roc_aucval:.3f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle=':', alpha=0.3)
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title(f'Validation ROC Curve (Alpha={n})', fontsize=14, fontweight='bold')
    plt.legend(loc='lower right', fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    plt.close()

    # Validation Confusion Matrix
    print('\nValidation Confusion Matrix:')
    ConfusionMat(xgb, X_train.iloc[:, 1:], y_train, validate_X.iloc[:, 1:],
                 validate_y, classes, n)

    # ============================================================
    # STEP 9: TEST SET (BLIND) EVALUATION
    # ============================================================
    print(f'\n--- Test Set (Blind Holdout) Evaluation ---')

    # Predictions
    pred = xgb.predict(X_test.iloc[:, 1:]).tolist()
    y_probs = xgb.predict_proba(X_test.iloc[:, 1:])[:, 1]

    # Store predictions
    predTEST.append(pred)
    actTEST.append(y_test.tolist())
    predTESTproba.append(y_probs.tolist())

    # Calculate metrics
    acc = accuracy_score(y_test, pred)
    preci = precision_score(y_test, pred, zero_division=0)
    recal = recall_score(y_test, pred, zero_division=0)
    f1test = f1_score(y_test, pred, zero_division=0)

    # Store test metrics
    accs.append(acc)
    prec.append(preci)
    rec.append(recal)
    f1scores_test.append(f1test)

    # Print test metrics
    print(f'Test Accuracy: {acc:.4f}')
    print(f'Test Precision: {preci:.4f}')
    print(f'Test Recall: {recal:.4f}')
    print(f'Test F1 Score: {f1test:.4f}')

    # Test AUC-ROC
    fpr, tpr, thresholds = roc_curve(y_test, y_probs)
    roc_auc = auc(fpr, tpr)
    aucscores.append(roc_auc)
    fprs.append(fpr)
    tprs.append(tpr)
    print(f'Test AUC Score: {roc_auc:.4f}')

    # Plot Test ROC Curve
    plt.style.use('ggplot')
    plt.figure(figsize=(10, 6), dpi=300)
    plt.plot(fpr, tpr, color='darkorange', lw=2, linestyle='--',
             label=f'Test AUC = {roc_auc:.3f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle=':', alpha=0.3)
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title(f'Test ROC Curve - Blind Holdout (Alpha={n})', fontsize=14, fontweight='bold')
    plt.legend(loc='lower right', fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    plt.close()

    # Test Confusion Matrix
    print('\nTest Confusion Matrix:')
    ConfusionMat(xgb, X_train.iloc[:, 1:], y_train, X_test.iloc[:, 1:],
                 y_test, classes, n)

    # ============================================================
    # STEP 10: CONFUSION MATRIX COMPONENTS (OPTIONAL DETAILED TRACKING)
    # ============================================================
    from sklearn.metrics import confusion_matrix

    # Calculate confusion matrix for test set
    cm_test = confusion_matrix(y_test, pred)
    if cm_test.shape == (2, 2):  # Binary classification
        tn, fp, fn, tp = cm_test.ravel()
        tns.append(tn)
        fps.append(fp)
        fns.append(fn)
        tps.append(tp)
        print(f'\nConfusion Matrix Components:')
        print(f'True Negatives: {tn}')
        print(f'False Positives: {fp}')
        print(f'False Negatives: {fn}')
        print(f'True Positives: {tp}')
    else:
        # For multi-class, store None or handle differently
        tns.append(None)
        fps.append(None)
        fns.append(None)
        tps.append(None)

    # ============================================================
    # STEP 11: SUMMARY FOR THIS ALPHA
    # ============================================================
    print(f'\n{"="*60}')
    print(f'Summary for Alpha {n}:')
    print(f'{"="*60}')
    print(f'Validation → Acc: {accval:.4f}, Prec: {precival:.4f}, Rec: {recalval:.4f}, F1: {f1val:.4f}, AUC: {roc_aucval:.4f}')
    print(f'Test (Blind) → Acc: {acc:.4f}, Prec: {preci:.4f}, Rec: {recal:.4f}, F1: {f1test:.4f}, AUC: {roc_auc:.4f}')
    print(f'{"="*60}\n')

# ============================================================
# STEP 12: FINAL AGGREGATED RESULTS ACROSS ALL ALPHAS
# ============================================================
print(f'\n{"#"*60}')
print(f'FINAL RESULTS ACROSS ALL ALPHAS')
print(f'{"#"*60}\n')

# Create results dataframe
results_df = pd.DataFrame({
    'Alpha': alphas,
    'Val_Accuracy': accsval,
    'Val_Precision': precval,
    'Val_Recall': recval,
    'Val_F1': f1scores_val,
    'Val_AUC': aucscoresval,
    'Test_Accuracy': accs,
    'Test_Precision': prec,
    'Test_Recall': rec,
    'Test_F1': f1scores_test,
    'Test_AUC': aucscores
})

print(results_df.to_string(index=False))

# Find best alpha based on validation AUC
valid_indices = [i for i, x in enumerate(aucscoresval) if x is not None]
if valid_indices:
    best_idx = max(valid_indices, key=lambda i: aucscoresval[i])
    print(f'\n🏆 Best Alpha: {alphas[best_idx]}')
    print(f'   Validation AUC: {aucscoresval[best_idx]:.4f}')
    print(f'   Test AUC: {aucscores[best_idx]:.4f}')
    print(f'   Test F1 Score: {f1scores_test[best_idx]:.4f}')

print(f'\n{"#"*60}\n')
```

```python
import plotly.graph_objects as go

# Example data
classes = ['Bad Ec', 'Good Ec']
actual_counts = [28, 11]  # Replace with actual class counts
predicted_counts = [26, 13]  # Replace with predicted class counts

# Create a grouped bar chart
fig = go.Figure()

# Add actual counts to the chart
fig.add_trace(go.Bar(
    x=classes,
    y=actual_counts,
    name='Actual Counts',
    marker_color='blue',
    text=[f"Actual = {val}" for val in actual_counts],
    textposition='auto'
))

# Add predicted counts to the chart
fig.add_trace(go.Bar(
    x=classes,
    y=predicted_counts,
    name='Predicted Counts',
    marker_color='orange',
    text=[f"Predicted = {val}" for val in predicted_counts],
    textposition='auto'
))

# Update layout for better appearance
fig.update_layout(
    title='Comparison of Actual and Predicted Good vs Bad Ec',
    xaxis_title='Class',
    yaxis_title='Counts',
    barmode='group',
#     legend=dict(title='Legend'),
    template='plotly_white'
)

# Show the chart
fig.show()
```
